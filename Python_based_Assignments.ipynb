```python
# 5. Prime Numbers in an Array
# Create an array with user-defined inputs. Use a for loop to fetch and print all prime numbers.

# Create an array with user-defined inputs
size = int(input("Enter the number of elements in the array: "))
array = []

# Input elements from the user
for _ in range(size):
    element = int(input("Enter an element: "))
    array.append(element)

# Print prime numbers from the array
print("Prime numbers in the array are:")
for num in array:
    if num < 2:  # Prime numbers are greater than or equal to 2
        continue
    is_prime = True  # Assume the number is prime
    for i in range(2, int(num ** 0.5) + 1):  # Check divisibility from 2 to sqrt(num)
        if num % i == 0:
            is_prime = False  # If divisible, it's not prime
            break
    if is_prime:
        print(num, end=' ')
```

```python
# 2. Even Numbers from an Array
# Write a function that takes an array as input and returns another array containing only the even numbers.

def get_even_numbers(arr):
    """
    Function to filter and return even numbers from the given array.
    
    Args:
    arr (list): The input array containing numbers.
    
    Returns:
    list: A list containing only even numbers.
    """
    even_numbers = []  # Initialize an empty list to store even numbers
    
    # Loop through each number in the array
    for num in arr:
        if num % 2 == 0:  # Check if the number is even
            even_numbers.append(num)  # Add the even number to the new list
    
    return even_numbers  # Return the list of even numbers

# Example usage
arr = [10, 23, 4, 26, 5, 75, 24, 54]
result = get_even_numbers(arr)
print("Even numbers in the array:", result)
```

```python
# 3. Array Intersection
# o Write a function that takes two arrays and prints the members of the first array that are also
# present in the second array. (Hint: Use membership comprehension.)

def array_intersection(arr1, arr2):
    """
    Function to find and print the elements of the first array that are also in the second array.
    
    Args:
    arr1 (list): The first input array.
    arr2 (list): The second input array.
    """
    # Using list comprehension to find common elements
    intersection = [i for i in arr1 if i in arr2]
    
    # Print the result
    print("Common elements:", intersection)

# Example usage
arr1 = [1, 2, 3, 4, 5]
arr2 = [3, 4, 5, 6, 7]
array_intersection(arr1, arr2)
```

```python
# Functions:
# 1. Write a function to find the factorial of a number using recursion.

def factorial_recursive(n):
    '''
    Funtion to calculate factorial recursive
    parameter : n is the integer to find out the factorial
    return: factorial of n
    '''
    if n==0 or n==1:
        return 1 #factorial of 0 or 1 is 1
    else:
        return n * factorial_recursive(n-1) #recursive case

#Example
num = 5
print(f"Factorial of {num} is: {factorial_recursive(num)}")
```

```python
#3. Create a function to check if a number is an Armstrong number.

def is_armstrong(number):
    """
    Function to check if a given number is an Armstrong number.
    :param number: Integer to check.
    :return: True if Armstrong, False otherwise.
    """
    # Convert the number to a string to access each digit
    num_str = str(number)
    n = len(num_str)  # Find the number of digits (order)
    
    # Calculate the sum of digits raised to the power n
    armstrong_sum = sum(int(digit) ** n for digit in num_str)
    
    # Check if the sum equals the original number
    return armstrong_sum == number

# Example usage:
num = 153
print(f"Is {num} an Armstrong number? {is_armstrong(num)}")

num = 123
print(f"Is {num} an Armstrong number? {is_armstrong(num)}")
```

```python
#4. Write a function that takes two strings and checks if one is an anagram of the other.


def is_anagram(str1, str2):
    # Step 1: Remove spaces and convert both strings to lowercase
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    
    # Step 2: Check if the lengths are the same
    if len(str1) != len(str2):
        return False
    
    # Step 3: Sort and compare both strings
    return sorted(str1) == sorted(str2)

# Example usage
print(is_anagram("listen", "silent"))   # Output: True
print(is_anagram("hello", "world"))     # Output: False
print(is_anagram("Dormitory", "Dirty room"))  # Output: True
```

```python
#5. Write a function to compute the sum of digits in a given number.

def sum_of_digits(number):
    # Step 1: Convert the number to a string
    number_str = str(abs(number))  # Use abs() to handle negative numbers
    
    # Step 2: Convert each character to an integer and sum them up
    total = sum(int(digit) for digit in number_str)
    
    return total

# Example usage
print(sum_of_digits(12345))   # Output: 15
print(sum_of_digits(9876))    # Output: 30
print(sum_of_digits(-123))    # Output: 6
```

```python
#2. Write a function to sum n NumPy arrays of the same dimensions.

import numpy as np

def sum_of_arrays(*arrays):
    # Initialize the result to a zero array of the same shape as the first array
    result = np.zeros_like(arrays[0]) # Assumes arrays have the same dimensions
    for array in arrays: # Loop through all the arrays and add each one to result
        result += array
    return(result) # Return the final summed array

#example

array1 =np.array([[1,2], [3,4]])
array2 = np.array([[5,6], [7,8]])
array3 = np.array([[9,10], [11, 12]])

result = sum_of_arrays(array1, array2, array3)
print(f"Sum of Arrays: \n{result}")
```

```python
#3. Write a function to extract the top-left N x M sub-matrix of a given 2D array.

import numpy as np

def extract_sub_matrix(array, N, M):
    # Extract the top-left N x M sub-matrix
    return array[0:N, 0:M]

# Example usage
array = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
N, M = 2, 2  # Specify the dimensions of the sub-matrix (2x2)
sub_matrix = extract_sub_matrix(array, N, M)
print(sub_matrix)
```

```python
#4. Write a function to extract the bottom-right N x M sub-matrix of a given 2D array.

import numpy as np

def extract_sub_matrix(array, N, M):
    # Extract the top-left N x M sub-matrix
    return array[-N:, -M:]

# Example usage
array = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
N, M = 2, 2  # Specify the dimensions of the sub-matrix (2x2)
sub_matrix = extract_sub_matrix(array, N, M)
print(sub_matrix)
```

```python
# 5. Write a function to calculate the mean and standard deviation of a 1D NumPy array, returning the result as a dictionary


def calculate_mean_std(array):

    '''
    Function to calculate mean and standard deviation of a 1D numpy array

    Parameters - A 1D Numpy array with numeric values

    Returns : A dictionary with keys mean and standard deviation.
    '''
    mean_value = np.mean(array)
    std_value = np.std(array)
    return {'mean': mean_value, 'std': std_value}

#1D array
array = np.array([23, 45, 12, 20, 15])

result = calculate_mean_std(array)
print(result)
```

```python
# Python Functions and Pandas Operations

#1. Write a function to generate a Pandas Series for a given range (start, end) with default values (1, 10)

import pandas as pd

def generate_series(start=1, end=10):
    """
    Function to generate a Pandas Series for a given range.
    
    Parameters:
    start (int): The starting value of the range (default is 1).
    end (int): The ending value of the range (default is 10).
    
    Returns:
    pandas.Series: A Pandas Series object with values from start to end-1.
    """
    return pd.Series(range(start, end))

# Example usage
series_default = generate_series()  # Using default values
print("Series with default range (1 to 10):\n", series_default)

series_custom = generate_series(5, 16)  # Using custom range
print("\nSeries with custom range (5 to 15):\n", series_custom)
```

```python
# 2. Create a function that accepts keys (string list) and values (list of lists) and returns a Pandas DataFrame.

def create_dataframe(keys, values):

    """
    Function to create a Pandas DataFrame from a list of keys and values.
    
    Parameters:
    keys (list of str): Column names for the DataFrame.
    values (list of lists): Data for each column.
    
    Returns:
    pandas.DataFrame: A Pandas DataFrame with the given keys as column names.
    """
    # Create the DataFrame using the keys and values
    df = pd.DataFrame({keys[i]: values[i] for i in range(len(keys))})
    return df

# Example usage
keys = ['Name', 'Age', 'City']
values = [['Rehana', 'Lara', 'Freya'], [25, 30, 35], ['Hyderabad', 'Chennai', 'Mumbai']]
dataframe = create_dataframe(keys, values)
print("Generated DataFrame:\n", dataframe)
```

```python
#3. Write a function to concatenate two DataFrames and reset their indices.

import pandas as pd

def concatenate_and_reset(df1, df2):
    """
    Function to concatenate two DataFrames and reset their indices.
    
    Parameters:
    df1 (pandas.DataFrame): The first DataFrame.
    df2 (pandas.DataFrame): The second DataFrame.
    
    Returns:
    pandas.DataFrame: A concatenated DataFrame with reset indices.
    """
    # Concatenate the two DataFrames
    concatenated_df = pd.concat([df1, df2], ignore_index=True)
    
    return concatenated_df

# Example usage
data1 = {'Name': ['Rehana', 'Lara'], 'Age': [25, 30]}
data2 = {'Name': ['Erin', 'Freya'], 'Age': [35, 40]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

result_df = concatenate_and_reset(df1, df2)
print("Concatenated DataFrame with reset indices:\n", result_df)
```

```python
#1. Load the employee_attrition.csv file into a pandas DataFrame.

import pandas as pd
data = pd.read_csv("HR-Employee-Attrition.csv")

df = pd.DataFrame(data)

df.head(10)
```

```python
#2. Calculate the percentage of employees who left the company.

#df.columns

#To calculate the percentage of employees who left the company we need total_employess count

#First calculate total_employees

# Count the total number of employees
total_employees = df['EmployeeCount'].sum()

# Count the number of employees who left (Attrition = 'Yes')

employees_left = len(df[df['Attrition'] == 'Yes'])

# Calculate the percentage of employees who left
percentage_left = (employees_left / total_employees) * 100

print(f"Percentage of employees who left the company: {percentage_left}")


# len(df) gives the total number of employees.
# df[df['Attrition'] == 'Yes'] filters the dataset for employees who left.
```

```python
#3. Identify the department with the highest attrition rate.

#We need first how many employees are there in each department

# Count total employees in each department
total_employees_by_dept = df['Department'].value_counts()
print(f"Total Employees by each Department: \n{total_employees_by_dept}")

# Count employees who left in each department
employees_left_by_each_dept = df[df['Attrition'] == 'Yes']['Department'].value_counts()
print(f"\nEmployees Left in each Department:\n {employees_left_by_each_dept}")

#calcualte the rate of attrition
attrition_rate_by_dept = (employees_left_by_each_dept/total_employees_by_dept) * 100
print(f"\nAttrition rate by Department: \n {attrition_rate_by_dept}")

# Find department with highest attrition rate
highset_attrition_dept = attrition_rate_by_dept.idxmax()
highest_attrition_rate = attrition_rate_by_dept.max()
print(f"\n Department with Highest attrition rate:{highset_attrition_dept} ({highest_attrition_rate})")

# idxmax() returns the department name with the highest attrition rate.
# max() returns the highest attrition rate value.
```

```python
#4. Find the average years of experience for employees who left and those who stayed.

# Average years of experience for employees who left the company
avg_year_left = df[df['Attrition'] == 'Yes']['YearsAtCompany'].mean()

# Average years of experience for employees who stayed with the company
avg_year_stayed = df[df['Attrition'] == 'No']['YearsAtCompany'].mean()

print(f"Average years of experience for employees who left: {avg_year_left}")
print(f"Average years of experience for employees who stayed: {avg_year_stayed}")
```

```python
#5. Filter the dataset for employees with a salary_hike greater than 15% and check their attrition status.

filtered_salary_hike = df[df['PercentSalaryHike'] > 15]
filtered_salary_hike

# Check their attrition status
attrition_rate = filtered_salary_hike['Attrition'].value_counts()

print(f"attrition_rate: {attrition_rate}")
```

```python
#2. Plot the closing prices for a given stock over time.

# Filter the data
closing_prices = df['Close']

# Plot the data
plt.figure(figsize=(12,8))
plt.plot(df['Date'], closing_prices, 
         linestyle='-', linewidth=2, color='lightcoral', 
         label='Closing Prices', alpha=0.8)

# Title and labels
plt.title("Closing Prices for a Given Stock Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Prices")

# Add grid with proper alpha value
#plt.grid(True, alpha=0.8)
plt.xticks(rotation =70)

# Show the plot
plt.legend()
plt.tight_layout() # Automatically adjusts the layout to prevent overlapping.
plt.show()
```

```python
#3. Calculate the daily percentage change for the stock prices.

df['daily_pct_change'] = df['Close'].pct_change() * 100 #Multiply with 100 to get the percent

# Display the first few rows 
df[['Date', 'Close', 'daily_pct_change']].head()
```

```python
#4. Identify the highest and lowest closing prices for each stock.

# Find the highest closing price
highest_closing_price = df['Close'].max()

#Finding the lowest closing price

lowest_closing_price = df['Close'].min()

print(f"Highest Closing Price for each stock: {highest_closing_price}")
print(f"Lowest Closing Price for each stock: {lowest_closing_price}")
```

```python
#5. Create a new column named moving_avg that calculates the 7-day moving average of the closing_price.

import pandas as pd

df['moving_average'] = df['Close'].rolling(window = 7).mean()

# Display the first few rows to check the new column
df.head(10)


# Why do we calculate the Moving Average in stock analysis?
# The moving average is a statistical tool used to smooth out short-term fluctuations in data and highlight longer-term trends.
# In stock analysis, moving averages help analysts and traders identify patterns, trends, and possible future price movements.
```

```python
#6. Find the correlation between the opening and closing prices.

correlation = df['Open'].corr(df['Close'])

print(f"Correlation between the opening and closing prices: {correlation:.2f}")


# Understanding the Output:
# Correlation Coefficient (r) ranges from -1 to 1:

# r = 1: Perfect positive correlation (both variables move in the same direction).
# r = -1: Perfect negative correlation (one variable increases while the other decreases).
# r = 0: No correlation (no linear relationship between the variables).
# A high positive correlation (close to 1) indicates that opening prices 
# and closing prices are closely related, meaning they generally move in the same direction.


#since in our output the correlation coefficient is 1 means there is a perfect positive correlation and both variables move in the same direction
```

# 4. Weather Data Analysis

```python
#1. Load the weather_data.csv file into a pandas DataFrame.

data = pd.read_csv("seattle-weather.csv")

#Creating a DataFrame

df = pd.DataFrame(data)

df.tail()

# Convert 'date' column to datetime format

df['date'] = pd.to_datetime(df['date'])
df
```

```python
#2. Find the average temperature for each month.

#since my data set contains temp_max and temp_min columns , I will sum p these two colunmns and divide by 2 to get average


#creating new column for aveg temp

df['Avg_Temp'] = (df['temp_max'] + df['temp_min']) / 2

# Extract month from the date
df['month'] = df['date'].dt.month

#Finding the average temperature for each month.

monthly_avg_temp = df.groupby('month')['Avg_Temp'].mean()
print(f"Average Temperature for each month:\n{monthly_avg_temp}")
```

```python
#3. Calculate the total precipitation for the year

total_precip = df['precipitation'].sum()

print(f"Total Precipitation for the year: {total_precip}")
```

```python
#4. Identify the day with the highest temperature and display its details.

#Extract the day from the date

df['day'] = df['date'].dt.day 

df['day']

#Identify the day with the highest temperature

highest_tem = df['day'].idxmax()

print(f"Highest temperature : {highest_tem}")

highest_temp_day = df['day'].max()

print(f"The day with highest temperature: {highest_temp_day}")
```

```python
#5. Visualize the temperature trend over the months using a line chart.

monthly_avg_temp = df.groupby('month')['Avg_Temp'].mean()

plt.figure(figsize = (12,6))

plt.plot(monthly_avg_temp.index, monthly_avg_temp.values, linestyle = '--', linewidth = 2, color = 'green', label = 'Temperature trend', 
         marker = 'd')

# Find the month with the highest average temperature
max_temp = monthly_avg_temp.max()  # Get the highest temperature
max_temp_month = monthly_avg_temp.idxmax()  # Get the month of the highest temperature

plt.title("Temperature trend over the months")
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.legend()
plt.annotate("Peak", 
             xy=(max_temp_month, max_temp),  # Coordinates of the peak point
             xytext=(max_temp_month + 1, max_temp + 1),  # Position of the text
             arrowprops=dict(facecolor='black', arrowstyle='->'),  # Arrow pointing to the peak
             fontsize=12, color='red')  # Customize text appearance
plt.show()
```

```python
# 6. Create a new column named temperature_diff that calculates the difference between the high and low temperatures for each day.

df['temperature_diff'] = df['temp_max'] - df['temp_min']

df['temperature_diff']
```

```python
#7. Filter the dataset for days where the temperature difference exceeds 15 degrees.

df[df['temperature_diff'] >15]
```
